```python
#Input Process File
import pandas as pd
import numpy as np
import random
import os
```

```python
#demand

# Define the network IDs and their corresponding rough magnitudes
network_ids = ['A', 'B', 'C', 'D', 'E', 'F']
rough_magnitudes = [0, 50000000, 30000000, 10000000, 50000000, 20000000]

# Create a DataFrame to store the data
data = {'Network ID': [], 'Day': [], 'Values': []}

# Generate data for each network ID
for network_id, magnitude in zip(network_ids, rough_magnitudes):
    for day in range(1, 366):  # For each day in the year
        value = np.random.poisson(magnitude)  # Generate a random value based on the rough magnitude
        data['Network ID'].append(network_id)
        data['Day'].append(day)
        data['Values'].append(value)

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
csv_file_path = 'Case_Study/Inputs/h2_demand.csv'
df.to_csv(csv_file_path, index=False)

df
```

```python
network_ids = ['C', 'E']
rough_magnitudes = [30000000, 50000000]

# Define the days for winter and summer peaks
winter_days = list(range(1, 80)) + list(range(355, 366))  # Approx. Dec-Feb
summer_days = list(range(172, 266))  # Approx. June-Aug

# Create a DataFrame to store the data
data = {'Network ID': [], 'Day': [], 'Values': []}

# Generate data for each network ID
for network_id, annual_magnitude in zip(network_ids, rough_magnitudes):
    total_values = 0
    for day in range(1, 366):  # For each day in the year
        if network_id == 'C':
            if day in winter_days:
                value = np.random.poisson(annual_magnitude / 60)  # Increase peak values in winter
            else:
                value = np.random.poisson(annual_magnitude / 365)
        elif network_id == 'E':
            if day in summer_days:
                value = np.random.poisson(annual_magnitude / 60)  # Increase peak values in summer
            else:
                value = np.random.poisson(annual_magnitude / 365)
        
        data['Network ID'].append(network_id)
        data['Day'].append(day)
        data['Values'].append(value)
        total_values += value
    
    # Adjust values to match the rough annual total
    factor = annual_magnitude / total_values
    data['Values'][-365:] = [round(v * factor) for v in data['Values'][-365:]]
df = pd.DataFrame(data)
csv_file_path = 'Case_Study/Inputs/h2_demand_CE.csv'
# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)
```
