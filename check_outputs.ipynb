```python
import pandas as pd
import numpy as np
import os
from glob import glob
import random

from Model_Functions_v2 import plot
```

```python
def comp_df(df1,df2,col=None):
    if col is not None:
        r = pd.concat([df1[col],df2[col]],axis=1)
        r.columns = ['{c}_{n}'.format(c=col,n=num) for num in range(1,3)]
    else:
        r = pd.concat([df1,df2],axis=1)
        r.columns = ['{c}_1'.format(c=df_col) for df_col in df1.columns] + ['{c}_2'.format(c=df_col) for df_col in df2.columns]
    return r
```

```python
def abs_max(data):
    return max([abs(i) for i in data])
```

```python
dir1 = 'Test Cases/test_case_recast_elec_prod2/'
dir2 = 'Test Cases/test_case_grid/'

p1 = [os.path.normpath(f) for f in glob(dir1+'*csv')]
p2 = [os.path.normpath(f) for f in glob(dir2+'*csv')]

d1 = {f.split('\\')[-1].split('.')[0]: pd.read_csv(f,index_col=0) for f in p1}
d2 = {f.split('\\')[-1].split('.')[0]: pd.read_csv(f,index_col=0) for f in p2}
```

```python
for d in d1:
    if d.startswith('Storage') or ('raw' in d):
        d1[d].set_index(d1[d].columns[0],append=True,inplace=True)

for d in d2:
    if d.startswith('Storage') or ('raw' in d):
        d2[d].set_index(d2[d].columns[0],append=True,inplace=True)
```

```python
key = 'Link_Flow'

sub_key = None

plot_all = False

if sub_key == None:
    to_plot1 = d1[key]
    to_plot2 = d2[key]
else:
    to_plot1 = d1[key].xs(sub_key)
    to_plot2 = d2[key].xs(sub_key)

if plot_all:
    col = None

else:
    col = d1[key].columns[random.randrange(0,len(d1[key].columns))]
    while col in checked_cols or (abs_max(to_plot1[col]) < 0.1 and abs_max(to_plot2[col]) < 0.1):
        col = d1[key].columns[random.randrange(0,len(d2[key].columns))]

    checked_cols.add(col)

plot(comp_df(to_plot1,to_plot2,col=col))
```
